#### pandas 메소드가 아닌 메소드로 agg()와 groupby() 메소드 조합

- agg() 와 aggregate() 메소드는 동일한 메소드
1. numpy.mean, numpy.sum 등의 메소드를 agg() 에 매개변수로 전달하여 수행 가능함.
2.  정의함수를 매개변수로 전달하여 수행 가능 함.

##### 사용자 함수 적용 예

```python
def my_mean(values):
    n=len(values)
    sum = 0
    for value in values:
        sum += value
    return sum / n

agg_my_mean = df.groupby('year')['lifeExp'].agg(my_mean)
agg_my_mean
```

```python
def my_mean_diff(values, diff_value):
    n = len(values)
    sum1 = 0
    for value in values:
        sum1 += value
    mean = sum1/ n
    return (mean - diff_value)
```

'''
- pd.concat(  
    objs: 'Iterable[Series | DataFrame] | Mapping[HashableT, Series | DataFrame]'
      *,
    axis: 'Axis' =\ 0,
    join: 'str' = 'outer',
    ignore_index: 'bool' = F                               * ignore_index = True 인경우 열 이름이 중복되지 않도록 설정가능함.alse,
    keys: 'Iterable[Hashable] | None' = None,
    levels=None,
    names: 'list[HashableT] | None' = None,
    verify_integrity: 'bool' = False,
    sort: 'bool' = False,
    copy: 'bool | None' = None,
) -> 'DataFra

HashableT는 pandas 라이브러리의 concat 함수의 타입 힌트에서 사용된 제네릭 타입 변수입니다. Hashable이라는 것은 파이썬 객체가 해시 가능(immutable) 하다는 것을 의미합니다. 해시 가능한 객체는 해시 테이블(예: 딕셔너리)의 키로 사용할 수 있습니다.

HashableT의 의미는 다음과 같습니다:

Hashable: 해당 객체가 해시 가능함을 나타냅니다. 즉, 객체는 변경 불가능하고, 동일한 객체는 동일한 해시 값을 갖습니다. 예를 들어, 문자열, 숫자, 튜플 등은 해시 가능 객체입니다.

T: 제네릭 타입 변수(T)를 나타내며, 이는 여러 타입으로 대체될 수 있습니다.

따라서 HashableT는 해시 가능 타입 변수를 의미하며, 이는 Mapping[HashableT, Series | DataFrame] 타입 힌트에서 사용됩니다. 이는 pandas.concat 함수에서 사용하는 키가 해시 가능한 타입이어야 한다는 것을 나타냅니다.

### 인덱스나 열 이름이 다른 데이터 연결하기

```python
df1.columns = ['A', 'B', 'C', 'D']
df2.columns = ['E', 'F','G', 'H']
df3.columns = ['A', 'C', 'F', 'H']
```

* concat() 함수를 사용하여 데이터 연결시 row_concat  예와 같이 결측값이 발생하게 되고 NaN 으로 채우게 됨.

```python
import pandas as pd

df6 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df7 = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})
df6
result = pd.concat({'first': df6, 'second': df7})
print(result)
```

```python
# 이중 인덱스 데이터프레임의 데이터 접근 방법. 예) 첫번째 행 첫번째 열의 데이터 1 에 접근하는 방법
print(result.iloc[0,0])      # iloc[] 사용법
print(result.loc[('first',0), 'A'])
```

```python
# 데이터 프레임 연결시 필연적으로 발생되는 NaN(결측치) 이 발생하지 않도록 하기 위해서는
# 데이터 프레임의 공통되는 부분만 연결하면 됨. 따라서 다음과 같이 연결하여 공통의 행 열 만 합칠 수 있음.
print(pd.concat([df1, df3],join = 'inner') )   # join = 'inner' : 이 옵션을 넣을 경우 공통의 행 열만 합치게 됨.
print(pd.concat([df2, df3],join = 'inner') ) 
print(pd.concat([df1, df2],join = 'inner') ) 
```
